Task 1 – Unit Converter Program

```python
# ==========================================
# 4.1 Task 1 - Unit Converter
# ==========================================

def convert_units(conversion_type, value):
    """
    Converts units between length, weight, and volume.

    Parameters:
    conversion_type (str): Type of conversion
    value (float): Value to convert

    Returns:
    float: Converted value
    """

    if conversion_type == "m_to_ft":
        return value * 3.28084
    elif conversion_type == "ft_to_m":
        return value / 3.28084
    elif conversion_type == "kg_to_lbs":
        return value * 2.20462
    elif conversion_type == "lbs_to_kg":
        return value / 2.20462
    elif conversion_type == "l_to_gal":
        return value * 0.264172
    elif conversion_type == "gal_to_l":
        return value / 0.264172
    else:
        raise ValueError("Unsupported conversion type.")


print("Available conversions:")
print("m_to_ft, ft_to_m")
print("kg_to_lbs, lbs_to_kg")
print("l_to_gal, gal_to_l")

choice = input("Enter conversion type: ")

try:
    value = float(input("Enter value: "))
    result = convert_units(choice, value)
    print("Converted value:", result)

except ValueError as e:
    print("Error:", e)
```

```python
# ==========================================
# 4.1 Task 2 - Mathematical Operations on a List
# ==========================================

def list_math_operations():
    """
    Performs sum, average, maximum, or minimum
    on a user-provided list of numbers.

    Returns:
    float: Result of selected mathematical operation
    """

    print("Select an Operation:")
    print("1. Sum")
    print("2. Average")
    print("3. Maximum")
    print("4. Minimum")

    choice = input("Enter operation (1/2/3/4): ")
    user_input = input("Enter a list of numbers separated by spaces: ")

    try:
        # Convert user input into list of floats
        num_list = [float(x) for x in user_input.split()]

        if not num_list:
            raise ValueError("List cannot be empty.")

        if choice == '1':
            return sum(num_list)

        elif choice == '2':
            return sum(num_list) / len(num_list)

        elif choice == '3':
            return max(num_list)

        elif choice == '4':
            return min(num_list)

        else:
            print("Invalid operation choice.")
            return None

    except ValueError as e:
        print("Error:", e)
        return None


# Call the function and print result
result = list_math_operations()

if result is not None:
    print("Result:", result)
```

Exercises on List Manipulation

```python
# ==========================================
# 4.2 Exercise on Lists
# List Slicing Operations
# ==========================================

# 1. Extract Every Other Element
def extract_every_other(lst):
    """
    Returns every second element from the list.
    """
    return lst[::2]


# 2. Slice a Sublist
def get_sublist(lst, start, end):
    """
    Returns a sublist from index 'start' to 'end' (inclusive).
    """
    return lst[start:end+1]


# 3. Remove the First and Last Elements
def remove_first_last(lst):
    """
    Removes the first and last elements of a list.
    Returns empty list if length < 2.
    """
    if len(lst) < 2:
        return []
    return lst[1:-1]


# 4. Get the First n Elements
def get_first_n(lst, n):
    """
    Returns the first n elements of the list.
    """
    return lst[:n]


# 5. Extract Last n Elements
def get_last_n(lst, n):
    """
    Returns the last n elements of the list.
    """
    if n == 0:
        return []
    return lst[-n:]


# 6. Reverse Order Skipping One Element
def reverse_skip(lst):
    """
    Returns elements in reverse order,
    skipping one element in between.
    """
    return lst[-2::-2]


# -----------------------------
# Test Cases
# -----------------------------

sample_list = [1, 2, 3, 4, 5, 6]

print("Original List:", sample_list)
print("Extract every other:", extract_every_other(sample_list))
print("Sublist (1 to 4):", get_sublist(sample_list, 1, 4))
print("Remove first and last:", remove_first_last(sample_list))
print("First 3 elements:", get_first_n(sample_list, 3))
print("Last 2 elements:", get_last_n(sample_list, 2))
print("Reverse skip:", reverse_skip(sample_list))
```

Exercises on Nested Lists

```python
# 1. Flatten a Nested List (1-Level Depth)
def flatten(lst):
    flat_list = []
    for item in lst:
        if isinstance(item, list):
            flat_list.extend(item)
        else:
            flat_list.append(item)
    return flat_list

# 2. Accessing Nested List Elements based on indices list
def access_nested_element(lst, indices):
    result = lst
    try:
        for idx in indices:
            result = result[idx]
        return result
    except (IndexError, TypeError):
        return "Invalid indices for the given list."

# 3. Sum of All Elements in a Nested List (Regardless of Depth)
def sum_nested(lst):
    total = 0
    for item in lst:
        if isinstance(item, list):
            total += sum_nested(item)
        elif isinstance(item, (int, float)):
            total += item
    return total

# 6. Count Occurrences of an Element in a Nested List
def count_occurrences(lst, elem):
    count = 0
    for item in lst:
        if isinstance(item, list):
            count += count_occurrences(item, elem)
        elif item == elem:
            count += 1
    return count

# Test Cases
nested = [[1, 2], [3, [4, 5]], 6]
print("Flatten 1 level:", flatten([[1, 2], [3, 4], [5]]))
print("Access element at [1, 2]:", access_nested_element([[1, 2, 3], [4, 5, 6], [7, 8, 9]], [1, 2]))
print("Sum of nested:", sum_nested(nested))
print("Count occurrences of 2:", count_occurrences([[1, 2], [2, 3], [2, 4]], 2))
```

Problem 4: Matrix Operations

```python
# Import Numpy
import numpy as np

# -----------------------------
# PART 1: Matrix Operations
# -----------------------------
A = np.array([[3, 4],
              [7, 8]])
B = np.array([[5, 3],
              [2, 1]])

print("Matrix A:\n", A)
print("Matrix B:\n", B)

# 1️⃣ Prove A * A^-1 = I
A_inv = np.linalg.inv(A)
print("\nInverse of A (A^-1):\n", A_inv)

I = np.dot(A, A_inv)
print("\nA * A^-1 =\n", I)

# 2️⃣ Prove AB != BA
AB = np.dot(A, B)
BA = np.dot(B, A)

print("\nA*B =\n", AB)
print("B*A =\n", BA)
print("Is A*B equal to B*A? ", np.array_equal(AB, BA))

# 3️⃣ Prove (AB)^T = B^T * A^T
AB_T = AB.T
BT_AT = np.dot(B.T, A.T)

print("\n(AB)^T =\n", AB_T)
print("B^T * A^T =\n", BT_AT)
print("Does (AB)^T equal B^T * A^T? ", np.array_equal(AB_T, BT_AT))

# -----------------------------
# PART 2: Solve Linear Equations using Inverse
# -----------------------------
# Equations:
# 2x - 3y + z = -1
# x - y + 2z = -3
# 3x + y - z = 9

# Coefficient matrix
A_eq = np.array([[2, -3, 1],
                 [1, -1, 2],
                 [3, 1, -1]])

# Constants vector
B_eq = np.array([-1, -3, 9])

# Solve using inverse
X_inv = np.dot(np.linalg.inv(A_eq), B_eq)
print("\nSolution using A^-1 * B:", X_inv)

# Solve using np.linalg.solve (recommended)
X_solve = np.linalg.solve(A_eq, B_eq)
print("Solution using np.linalg.solve:", X_solve)
```
